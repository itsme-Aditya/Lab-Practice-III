```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x + 3)**2

def f_prime(x):
    return 2 * (x + 3)
```

```python
def gradient_descent(learning_rate=0.1, epochs=25, start_x=2):
    x = start_x
    x_history = [x]
    y_history = [f(x)]
    
    for _ in range(epochs):
        slope = f_prime(x)
        x = x - learning_rate * slope
        x_history.append(x)
        y_history.append(f(x))
        
    return x, x_history, y_history
```

```python
min_x, x_hist, y_hist = gradient_descent(learning_rate=0.1, epochs=25, start_x=2)
```

```python
print(f"Local minimum occurs at x ≈ {min_x:.4f}")
```

```python
x_vals = np.linspace(-6, 4, 200)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label='y = (x + 3)^2')
plt.scatter(x_hist, y_hist, color='red', label='Gradient descent steps')
plt.title('Gradient Descent Optimization')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
